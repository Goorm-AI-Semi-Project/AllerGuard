```python
import io
import os
from google.cloud import vision

# 본인의 서비스 계정 키 파일 경로로 설정하세요.
key_path = r"C:\Users\MYNOTE\Desktop\새 폴더\ocr-project-470906-7ffeebabebeb09.json"
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = key_path
# -----------------------------------------

# 1. 본인이 확인하고 싶은 알레르기 성분 목록을 정의합니다.
# 여기에 원하는 성분을 자유롭게 추가하거나 제거하세요.
ALLERGY_LIST = [
    '메밀', '밀', '대두', '잣', '호두', '땅콩', '아몬드',
    '복숭아', '토마토', '우유', '새우', '고등어', '오징어',
    '게', '조개류', '굴', '전복', '홍합', '돼지고기', '닭고기',
    '쇠고기', '난류', '계란', '알류'
]

def detect_text(path):
    """
    이미지 파일에서 텍스트를 감지하여 문자열로 반환합니다.
    """
    client = vision.ImageAnnotatorClient()

    if not os.path.exists(path):
        return f"오류: 파일을 찾을 수 없습니다. 경로: {path}"

    try:
        with io.open(path, 'rb') as image_file:
            content = image_file.read()

        image = vision.Image(content=content)
        response = client.text_detection(image=image)

        if response.error.message:
            raise Exception(
                '{}\nFor more info on error messages, check: '
                'https://cloud.google.com/apis/design/errors'.format(
                    response.error.message))

        texts = response.text_annotations
        if texts:
            return texts[0].description
        else:
            return '텍스트를 찾을 수 없습니다.'
    except Exception as e:
        return f"오류 발생: {e}"


def process_folder(folder_path):
    """
    지정된 폴더 내의 모든 .jpg 파일에서 텍스트를 추출하고,
    미리 정의된 알레르기 목록과 비교하여 결과를 출력합니다.
    """
    print(f"'{folder_path}' 폴더의 모든 JPG 파일 처리를 시작합니다.\n")
    
    for filename in os.listdir(folder_path):
        if filename.lower().endswith('.jpg'):
            image_path = os.path.join(folder_path, filename)
            
            # 이미지에서 전체 텍스트를 추출합니다.
            extracted_text = detect_text(image_path)
            
            # 현재 파일에서 발견된 알레르기 성분을 저장할 리스트
            found_allergens = []

            # 2. 추출된 텍스트에 알레르기 목록의 각 성분이 포함되어 있는지 확인합니다.
            if '오류' not in extracted_text:
                for allergen in ALLERGY_LIST:
                    if allergen in extracted_text:
                        found_allergens.append(allergen)
            
            # 3. 형식에 맞게 결과를 출력합니다.
            print(f"========================================")
            print(f"■ 파일명: {filename}")
            print(f"----------------------------------------")

            if found_allergens:
                # join()을 사용하여 리스트를 예쁘게 출력합니다. 예: '우유, 대두, 밀'
                detected_list_str = ', '.join(found_allergens)
                print(f"▶ 검출된 알레르기 성분: {detected_list_str} ◀")
            else:
                print("▶ 알레르기 성분이 검출되지 않았습니다.")
            
            print(f"----------------------------------------")
            print(f"[전체 추출 텍스트]\n{extracted_text.strip()}")
            print(f"========================================\n")


# -----------------------------------------
#       *** 실행 부분 ***
# -----------------------------------------
YOUR_FOLDER_PATH = r'C:\Users\MYNOTE\AllerGuard\Data' 
process_folder(YOUR_FOLDER_PATH)
```

```python
template_for_extract = """
주어진 텍스트 내용을 확인해서 원재료명과 혼입가능 혹은 같은제조시설에서 사용되는 재료를 분류해주세요.
원재료명 내에서 중복되는 항목은 제거하고 해당되는 내용이 없으면 '없음' 이라고 답해주세요
혼입가능 혹은 같은제조시설에서 사용되는 재료중 중복되는 항목은 제거하고 해당되는 내용이 없으면 '없음' 이라고 답해주세요
재료한개당 한줄에 작성해주세요.
다음 JSON 포맷 에 맞게 정리해주세요.

----
원문 텍스트:
{raw_text}
----

----
[JSON 포맷]
{{
  "원재료명": [],
  "혼입가능": []
}}

혼입가능,같은제조시설:
----

#Answer:
"""
```

```python
# 텍스트 파일의 경로
file_path = "./data/김광무_100_ocr.txt"

# 파일 내용 읽기
with open(file_path, 'r', encoding='utf-8') as f:
    raw_text = f.read()

print("raw_text=",raw_text)
```

```python
prompt_for_extract = PromptTemplate.from_template(template_for_extract)
prompt_for_extract
```

```python
# chain 생성
chain_for_extract = prompt_for_extract | llm
```

```python
res = chain_for_extract.invoke({"raw_text":raw_text})
```

```python
# template 정의. {country}는 변수로, 이후에 값이 들어갈 자리를 의미
template_for_categorize = """{ingredients}은/는 다음 19가지 카테고리중 어디에 포함되나요? 포함된다면 카테고리 이름을 알려주고 어디에도 포함되지 않으면 '그외' 라고 응답해주고 앞에서 지시한 카테고리 이름이나 '그외'라는 단어 이외에는 아무것도 답하지 마세요.
- 카테고리:
난류(계란류,가금류)
소고기
돼지고기
닭고기
새우
게
오징어
고등어
조개류(굴,전복,홍합포함)
우유
땅콩
호두
잣
대두
복숭아
토마토
밀
메밀
아황산류
"""

# from_template 메소드를 이용하여 PromptTemplate 객체 생성
prompt_for_categorize = PromptTemplate.from_template(template_for_categorize)
prompt_for_categorize
```

```python
# chain 생성
chain_for_categorize = prompt_for_categorize | llm
```

```python
import json
# json.loads() 함수를 이용해 문자열을 파이썬 딕셔너리로 변환
json_data = json.loads(res.content)

# 딕셔너리의 키-값 쌍을 반복합니다.
for key, value in json_data.items():
    # '없음'을 포함하지 않는 재료만 필터링합니다.
    filtered_ingredients = [ingredient for ingredient in value if ingredient != "없음"]

    # 만약 필터링된 재료 리스트가 비어있지 않다면 출력합니다.
    if filtered_ingredients:
        print(f"* {key} : ")
        for ingredient in filtered_ingredients:
            print(f"  - {ingredient} : {chain_for_categorize.invoke(ingredient).content}")
```

```python
import os
from langchain_community.vectorstores import FAISS
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OpenAIEmbeddings # 또는 다른 임베딩 모델
from langchain_google_community import GoogleSearchAPIWrapper
from langchain_core.documents import Document

# 1. API 키 설정 (Google Cloud Console에서 발급받은 키와 CSE ID를 사용)
# os.environ["GOOGLE_API_KEY"] = "YOUR_GOOGLE_API_KEY"
# os.environ["GOOGLE_CSE_ID"] = "YOUR_GOOGLE_CSE_ID"

# 2. Google Search API 래퍼 초기화
# 'k' 인수를 제거하고 기본 초기화
search = GoogleSearchAPIWrapper()

# 3. 검색어 설정 및 데이터 조회
query = "치자청색소는 알러지 유발물질인가요?"
# 'num_results' 인자를 사용하여 결과 개수를 명시
search_results = search.results(query, num_results=5)

# 4. LangChain Document 객체 생성
documents = []
for result in search_results:
    doc = Document(
        page_content=result['snippet'],
        metadata={"source": result['link'], "title": result['title']}
    )
    documents.append(doc)

# 5. 임베딩 모델 로드 (FAISS와 함께 사용할 임베딩 모델)
# FAISS는 벡터를 저장하므로, 벡터를 생성해주는 임베딩 모델이 필요합니다.
# 이 예시에서는 OpenAIEmbeddings를 사용하지만, 다른 모델도 가능합니다.
# from langchain_google_genai import GoogleGenerativeAIEmbeddings
# embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
embeddings = OpenAIEmbeddings()

# 6. FAISS 벡터스토어 생성 및 저장
# FAISS.from_documents()는 문서 리스트를 받아 임베딩하고 FAISS 인덱스를 생성합니다.
faiss_db = FAISS.from_documents(documents, embeddings)

# 7. 생성된 FAISS 인덱스를 파일로 저장
# "google_search_index"라는 디렉토리가 생성되고, 인덱스 파일이 저장됩니다.
faiss_db.save_local("aller_item_index")

print("Google Search 결과가 FAISS DB에 출처 URL과 함께 성공적으로 저장되었습니다.")
```

```python
faiss_aller_db = None

def tav_search(name):
    global faiss_aller_db
    
    query = f"{name}는 알러지 유발물질인가요?"
    search_result = search.invoke(query)
    
    documents = []
    for result in search_result:
        doc = Document(
            # 'snippet' 대신 'content'를 사용해 더 긴 내용을 담을 수 있습니다.
            page_content=result['content'],
            metadata={"source": result['url'], "title": result['title']}
        )
        print("name=",name,"doc=",doc)
        documents.append(doc)
        
    if faiss_aller_db is None:
        faiss_aller_db = FAISS.from_documents(documents, embeddings)
    else:
        faiss_aller_db.add_documents(documents)
```

```python
# 문서를 예쁘게 출력하기 위한 도우미 함수
def pretty_print_docs(docs):
    print(
        f"\n{'-' * 100}\n".join(
            [f"문서 {i+1}:\n\n" + d.page_content for i, d in enumerate(docs)]
        )
    )
```

```python
from langchain_teddynote.document_compressors import LLMChainExtractor
from langchain.retrievers import ContextualCompressionRetriever

# loaded_faiss_db를 as_retriever()를 사용하여 Runnable 타입으로 변환
base_retriever = loaded_faiss_db.as_retriever()

# LLM을 사용하여 문서 압축기 생성
compressor = LLMChainExtractor.from_llm(llm)
compression_retriever = ContextualCompressionRetriever(
    # 문서 압축기와 리트리버를 사용하여 컨텍스트 압축 리트리버 생성
    base_compressor=compressor,
    base_retriever=base_retriever,
)

compressed_docs = (
    compression_retriever.invoke(  # 컨텍스트 압축 리트리버를 사용하여 관련 문서 검색
        "아황산류의 유사어에 대해서 알려줘"
    )
)
pretty_print_docs(compressed_docs)  # 검색된 문서를 예쁘게 출력
```

```python
# 1. 검색어가 포함된 데이터가 '있는' 경우
query_with_result = "치자청색소의 유해성에 대해 알려줘"
compressed_docs = compression_retriever.invoke(query_with_result)

if compressed_docs:
    print(f"'{query_with_result}'에 대한 검색 결과를 찾았습니다.", compressed_docs)
    # pretty_print_docs(compressed_docs) # 실제 문서 출력
else:
    print(f"'{query_with_result}'에 대한 검색 결과가 없습니다.")

print("\n---")

# 2. 검색어가 포함된 데이터가 '없는' 경우
query_without_result = "바나나의 제조공정에 대해 알려줘" # 이 데이터는 DB에 없다고 가정
compressed_docs_no_result = compression_retriever.invoke(query_without_result)

if compressed_docs_no_result:
    print(f"'{query_without_result}'에 대한 검색 결과를 찾았습니다.", compressed_docs_no_result)
    # pretty_print_docs(compressed_docs_no_result)
else:
    print(f"'{query_without_result}'에 대한 검색 결과가 없습니다.")
```
